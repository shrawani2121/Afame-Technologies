```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
plt.hist(df['Sales'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x='Quantity',data=df)
plt.show()
```

```python
sns.countplot(x='State',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# State with highest sales
sales_by_state = df.groupby('State')['Sales'].sum()
highest_sales_state = sales_by_state.idxmax()
plt.figure(figsize=(10, 6))
plt.bar(sales_by_state.index, sales_by_state.values, color='orange')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.title('Total Sales by State')
plt.bar(highest_sales_state, sales_by_state[highest_sales_state], color='red', label=f'Highest Sales ({highest_sales_state})')
plt.show()
```

```python
# which Category has highest sales
sales_by_state = df.groupby('Category')['Sales'].sum()
highest_sales_state = sales_by_state.idxmax()
plt.figure(figsize=(10, 6))
plt.bar(sales_by_state.index, sales_by_state.values, color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.title('Total Sales by Category')
plt.bar(highest_sales_state, sales_by_state[highest_sales_state], color='green', label=f'Highest Sales ({highest_sales_state})')
plt.show()
```

```python
# which sub category has highest sales
sales_by_state = df.groupby('Sub-Category')['Sales'].sum()
highest_sales_state = sales_by_state.idxmax()
plt.figure(figsize=(10, 6))
plt.bar(sales_by_state.index, sales_by_state.values, color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.title('Total Sales by Sub-Category')
plt.bar(highest_sales_state, sales_by_state[highest_sales_state], color='red', label=f'Highest Sales ({highest_sales_state})')
plt.show()
```

```python
df.columns
```

```python
#which city in each state has highest sales
highest_sales_per_state =df.groupby('State').apply(lambda x: x.loc[x['Sales'].idxmax()])
print(highest_sales_per_state[[ 'City']])
```

```python
# Top 5 States with highest sales
sales_by_state = df.groupby('State')['Sales'].sum()
sorted_states = sales_by_state.sort_values(ascending=False)
top_5_states = sorted_states.head(5)
plt.figure(figsize=(3,3))
plt.pie(top_5_states, labels=top_5_states.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 States with Highest Sales')
plt.axis('equal')
plt.show()
```

```python
#Top 10 higeshst selling products
sales_by_product = df.groupby('Product Name')['Sales'].sum()
sorted_products = sales_by_product.sort_values(ascending=False)
top_10_products = sorted_products.head(10)
plt.figure(figsize=(10, 6))
top_10_products.plot(kind='bar', color='skyblue')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Top 10 Products with Highest Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# which segment has highest sales
sales_by_segment =df.groupby('Segment')['Sales'].sum()
highest_sales_segment = sales_by_segment.idxmax()
plt.figure(figsize=(10, 6))
sales_by_segment.plot(kind='bar', color='skyblue')
plt.xlabel('Segment')
plt.ylabel('Total Sales')
plt.title('Total Sales by Segment')
plt.bar(highest_sales_segment, sales_by_segment[highest_sales_segment], color='red', label=f'Highest Sales ({highest_sales_segment})')
plt.show()
```

```python
# Which region has highest sales
sales_by_region =df.groupby('Region')['Sales'].sum()
highest_sales_region = sales_by_region.idxmax()
plt.figure(figsize=(10, 6))
sales_by_region.plot(kind='bar', color='brown')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Total Sales by Region')
plt.bar(highest_sales_region, sales_by_region[highest_sales_region], color='grey', label=f'Highest Sales ({highest_sales_region})')
plt.show()
```

```python
# What Quantity has highest discount
discount_by_quantity =df.groupby('Quantity')['Discount'].sum()
highest_discount_quantity = discount_by_quantity.idxmax()
plt.figure(figsize=(10, 6))
discount_by_quantity.plot(kind='line', color='skyblue')
plt.xlabel('Quantity')
plt.ylabel('Total Discount')
plt.title('Total Discount by Quantity')
plt.plot(highest_discount_quantity, discount_by_quantity[highest_discount_quantity], marker='o', markersize=8, color='red', label=f'Highest Discount ({highest_discount_quantity})')
plt.show()
```

```python
# Which category has highest profit
profit_by_category =df.groupby('Category')['Profit'].sum()
highest_profit_category = profit_by_category.idxmax()
plt.figure(figsize=(3,3))
plt.pie(profit_by_category, labels=profit_by_category.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Profit by Category')
plt.axis('equal')
plt.show()
```

```python
# Top 5 products with highest discount
sales_by_product =df.groupby('Product Name')['Sales'].sum()
sorted_products = sales_by_product.sort_values(ascending=False)
top_5_products = sorted_products.head(5)
print(top_5_products)
```

```python
# Top 10 product with most shipping cost
shipping_cost_by_product =df.groupby('Product Name')['Shipping Cost'].sum()
sorted_products = shipping_cost_by_product.sort_values(ascending=False)
top_10_products = sorted_products.head(10)
plt.figure(figsize=(12, 6))
top_10_products.plot(kind='bar', color='orange')

plt.xlabel('Product')
plt.ylabel('Total Shipping Cost')
plt.title('Top 10 Products with Highest Shipping Cost')
plt.xticks(rotation=90)
plt.show()
```
